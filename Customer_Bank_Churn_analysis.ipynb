## What factors influence customer churn (Exited)?

This question would involve analyzing the relationship between customer characteristics (e.g., CreditScore, Age, Geography, Balance, NumOfProducts, etc.) and whether a customer has churned (Exited = 1 or 0). The goal would be to identify key predictors of churn.


```python
# Drop columns that are not needed
df_filtered = df.drop(columns=['CustomerId', 'Surname', 'Age_range', 'Salary_range'])

# Label encoding for binary columns like 'Gender', 'HasCrCard', 'IsActiveMember'
# df_filtered['Gender'] = df_filtered['Gender'].map({'Female': 0, 'Male': 1})
df_filtered['HasCrCard'] = df_filtered['HasCrCard'].map({'No': 0, 'Yes': 1})

# One-hot encoding for 'Geography and Gender'
df_filtered = pd.get_dummies(df_filtered, columns=['Geography', 'Gender'])


# Display the final DataFrame
df_filtered[['Geography_France', 'Geography_Germany', 'Geography_Spain']] = df_filtered[['Geography_France', 'Geography_Germany', 'Geography_Spain']].astype(int)
df_filtered[['Gender_Male', 'Gender_Female']] = df_filtered[['Gender_Male', 'Gender_Female']].astype(int)
df_filtered
```

```python
# Compute the correlation matrix
correlation_matrix = df_filtered.corr()

# Plot a heatmap to visualize the correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Set the title and show the plot
plt.title('Correlation Heatmap of Customer Churn Factors')
plt.show()
```

Based on the heatmap showing the correlation between the **Exited** variable (customer churn) and other features, the following relationships have been identified:

 **Age**  
The strongest positive correlation with churn is seen with **Age** (0.29). This suggests that older customers are more likely to churn.

 **Balance**  
There is a mild positive correlation between **Balance** and churn (0.12). Customers with higher balances seem slightly more likely to churn.

 **Geography**  
- **Germany** has a positive correlation with churn (0.17), indicating that customers from Germany are more likely to churn.  
- **France** and **Spain** have negative correlations (-0.14 and -0.05, respectively), suggesting that customers from these countries are less likely to churn compared to Germany.

 **IsActiveMember**  
There is a negative correlation with churn (-0.16), indicating that active members are less likely to churn.

 **NumOfProducts**  
Shows a slight negative correlation (-0.05), suggesting that customers with more products tend to stay longer.

 **Gender**  
Being **female** has a slight positive correlation with churn (0.11), suggesting that women are marginally more likely to churn than men.

Other variables like **CreditScore**, **EstimatedSalary**, **Tenure**, and **HasCrCard** have relatively weak correlations with churn, implying that they are not strong predictors on their own.

---

 **Overall Summary**  
**Age**, **Balance**, **Geography (Germany)**, and **IsActiveMember** are the most significant factors affecting churn based on this heatmap.


## Multiple linear Regression analysis

```python
#training features except the Exited collumn 
x = df_filtered.drop(columns = "Exited")
x
```

```python
# dependent variable
y = df_filtered['Exited']
y
```

```python
#split the data to carry out t-test
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3, random_state=0)
```

```python
#Linear regression model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
#train our model on our training data
lr.fit(x_train, y_train)
```

```python
#get our y-intercept and coefficients
c = lr.intercept_
m = lr.coef_
print(f"Intercept:{c}\nCoefficients:{m}")
```

```python
#test if our model is working and get the predicted exited using our training set
y_pred_train = lr.predict(x_train)
y_pred_train
```

```python
#visualize the predicted values and our actual values
plt.scatter(y_train, y_pred_train)
plt.xlabel("Actual Exited Values (Trained Data")
plt.ylabel("Predcited Churn Values")
plt.show()
```

```python
#compute the r2 score of our tained data
from sklearn.metrics import r2_score
r2_score(y_train, y_pred_train)
```

## Interpretation of the R² (0.154)

The model explains about 15.4% of the variance in the target variable. In other words, only a small fraction of the changes in the churn status can be explained by the model. This confirms our earlier observation with the heat map that there are no significant predictors of our Exited collumn

```python
#test the model with our testing data
y_pred_test = lr.predict(x_test)
```

```python
#visualize the predicted values and our test data
plt.scatter(y_test, y_pred_test)
plt.xlabel("Actual Exited Values (unseen Data)")
plt.ylabel("Predcited Churn Values")
plt.show()
```

```python
#compute r2 score of test data
from sklearn.metrics import r2_score
r2_score(y_test, y_pred_test)
```

## Conclusion and Recomendation

Since Both the R2 values are relatively low and given the correlation matrix reveals a weak to moderate relationships between features and churn the model struggles to capture these weak patterns, leading to underfitting. To adrees this issues :
